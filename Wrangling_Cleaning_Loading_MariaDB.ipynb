# Introduce more impurity

## NOTE: For academic purpose, we will introduce impurity randomly into the data

### Duplicate rows

```python
# The proportion of rows to duplicate is 3%

proportion_to_duplicate = 0.03

# num_rows_to_duplicate indicates how many row indices should be randomly chosen.
# Want to return an integer because the index are not int and not float

num_rows_to_duplicate = int(len(data_m) * proportion_to_duplicate) 


# Randomly select the row indeces to duplicate:

# a) The fist argument of the random.choice method is specifying that the random selection should be made from the indices (row labels) of 
#    the DataFrame data_m
# b) The second argument of the random.choice method specifies the number of elements to randomly select, which is 3% of the data_m rows
# c) The third argument means that the same index cannot be chosen more than once, ensuring that all selected indices are unique

rows_to_duplicate = np.random.choice(data_m.index, size=num_rows_to_duplicate, replace=False)
#rows_to_duplicate

# Duplicate the selected rows indeces:

# The code below creates a new DataFrame called duplicated_rows that consists of specific rows from data_m, 
# determined by the indices provided in rows_to_duplicate

duplicated_rows = data_m.loc[rows_to_duplicate]
#duplicated_rows

# Append duplicated rows to the original DataFrame: Now we concatenate the duplicated rows to the data_m 
# dataframe in order to have a a dataframe with duplicated rows
# The data_world has 235 rows. However the data_with_duplicates has 242 rows. Therefore, it was added 7 more rows

data_with_duplicates = pd.concat([data_m, duplicated_rows])

data_with_duplicates
```

# Introduce Symbols

```python
# These data is relatively clean. Fpr academic rason, We will introduce some impurity randomly. 
# These will allow us to apply Pandas tools to clean data  
```

```python
# Symbols to be introduced
symbols = ["/", ".", "_"]

proportion_to_introduce_symbols = 0.03

# Number of symbols to introduce
num_symbols_to_introduce = int(len(data_with_duplicates) * proportion_to_introduce_symbols)

# Select random indices in the column where symbols will be introduced
random_indices = np.random.choice(data_with_duplicates.index, size=num_symbols_to_introduce, replace=False)

# Introduce symbols at the selected indices
for index in random_indices:
    random_symbol = np.random.choice(symbols)
    data_with_duplicates.at[index, "Country_area"] += random_symbol

data = data_with_duplicates
data
```
